```python
train["source"] = "train"
test["source"] = "test"
```

```python
data = pd.concat([train, test], ignore_index=True, axis=0)
```

```python
data.apply(lambda x: sum(x.isnull()))
data["Age"].value_counts()
```

```python
data.apply(lambda x: len(x.unique()))
```

```python
data.dtypes
```

```python
data["Age"] = data["Age"].replace({"26-35":0,"36-45":1,"18-25":2,"46-50":3,"51-55":4,"0-17":5,"55+":6})
```

```python
genderDict = {"M":0,"F":1}
data["Gender"] = data["Gender"].apply(lambda x: genderDict[x])
```

```python
citycatDict = {"A":0,"B":1,"C":2}
data["City_Category"] = data["City_Category"].apply(lambda x: citycatDict[x])
```

```python
data.dtypes
```

```python
data["Stay_In_Current_City_Years"].value_counts()
```

```python
stayDict = {"1":0,"2":1,"3":2,"4+":3,"0":4}
data["Stay_In_Current_City_Years"] = data["Stay_In_Current_City_Years"].apply(lambda x: stayDict[x])
```
